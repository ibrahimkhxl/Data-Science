### Machine learning models work very well for dataset having only numbers. But how do we handle text information in dataset? Simple approach is to use interger or label encoding but when categorical variables are nominal

 using simple label encoding can be problematic. One hot encoding is the technique that can help in this situation. In this tutorial, we will use pandas get_dummies method to create dummy variables that allows us to perform one hot encoding on given dataset. Alternatively we can use sklearn.preprocessing OneHotEncoder as well to create dummy variables. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town,dtype=int)
dummies
```

```python
merged = pd.concat([df,dummies],axis = 1)
merged
```

```python
merged.drop('town',axis = 1,inplace = True)
```

we have to drop one dummie variable column \
for example if we have 3 dummie variable column we need to have 2\
just like if we have 5 dummie column we need to have 4

```python
merged
```

```python
final = merged.drop('west windsor',axis = 1)
```

```python
final
```

```python
reg = linear_model.LinearRegression()
```

```python
X = final.drop('price',axis = 1)
X
```

```python
Y = final.price
Y
```

```python
reg.fit(X,Y)
```

```python
# predict price of area 2800 in robinsville
reg.predict([[2800,0,1]])
```

```python
# predict price of area 3400 in west windsor
reg.predict([[3400,0,0]])
```

```python
# predict price of area 3000 in monroe township
reg.predict([[3000,1,0]])
```

### To check the accuracy of our model

```python
reg.score(X,Y)
```
