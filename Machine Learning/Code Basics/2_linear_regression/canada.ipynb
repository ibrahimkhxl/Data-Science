# Given in the dataset we have year and avg income in canada. we need to use linear model and finf the avg income in year 2020


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\ibrah\Data Science\Machine Learning\Code Basics\1_linear_regression\canada_per_capita_income.csv")
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Income")
plt.title(" Canada per capita income (US$)")
plt.scatter(df.year,df['per capita income (US$)'])
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
x = reg.predict([[2020]])
x
```

```python
print("The per capita income for Year 2020 is",x,"$")
```
