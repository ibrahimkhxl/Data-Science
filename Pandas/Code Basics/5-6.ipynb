# 6. Replace Method to deal with missing data

```python
import numpy as np
data = pd.read_csv(r"C:\Users\ibrah\OneDrive\Desktop\Workshop\Pandas_Codebasics\weather_data6.csv")
data
```

```python
new_data = data.replace(-99999, np.NaN)
new_data
```

```python
# data.loc[data.temperature, ['NaN_temp']] = data.temperature
# data.loc[data.temperature < -500, ['NaN_temp']] = np.NaN
# loc_data.replace(loc_data.windspeed ,0)
# loc_data
data['NaN_Temperature'] = data.iloc[:,2]
```

```python
data
```

```python
data.drop(columns = ['NaN_Temperature'],inplace = True)
```

```python
type(data.event[3].st.contains)
```

```python
data = data.replace(
{
    'temperature': -99999,
    'windspeed': -99999,
    'event': '0',
},np.NaN)
```

```python
data.isna().sum()
```

```python
 data['temperature'] = data.temperature.fillna(data.temperature.mean())
```

```python
 data['windspeed'] = data.windspeed.fillna(data.windspeed.mean())
```

```python
data.event= data.event.fillna(method = 'ffill')
```

```python
data
```

```python
data.loc[(data.temperature >= 31) & (data['event'] == 'Sunny'), ['Status']] = "Warm"
```

```python
data.loc[(data.temperature < 31) | (data['event'] != 'Sunny'), ['Status']] = "Cold"
```
