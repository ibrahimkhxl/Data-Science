Leagend

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")

plt.plot(days,max_t,label = "Max")
plt.plot(days,min_t,label = "Min")
plt.plot(days,avg_t,label = "Avg",linestyle = "dashed")

plt.legend(loc = "best",shadow = True,fontsize = "small")

plt.grid()
```

### T5

```python
# Histagram 
```

only 1D array and y axis contains the frequency of the array

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
```

here normal is 80-100 , pre diabetic is 100-125 and diabetic is 125 above

```python
plt.hist(blood_sugar, bins = [80,100,125,200],rwidth = 0.90,color = "red",orientation = "horizontal")
```

2 type of histogram side by side

```python
blood_sugar_man = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist([blood_sugar_man, blood_sugar_women],bins = [80,100,125,200],rwidth = 0.85,
         color = ["blue",'pink'],label = ["Men","Women"])
plt.legend()

plt.xlabel("Sugar level")
plt.ylabel("Patients")
plt.title("Blood sugar analysis")
```

### T6

Pie-Chart

```python
exp_val = [1400,600,300,410,250]
exp_lab = ["rent",'Food','Wifi','Car','Other']
```

```python
plt.pie(exp_val,labels = exp_lab)
plt.show()
```

```python
plt.pie(exp_val,labels = exp_lab,radius = 1.5,autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie(exp_val,labels = exp_lab,autopct = '%0.1f%%',explode = [0,0.2,0,0,0])
plt.show()

plt.savefig("pie.pdf",bbox_inches = "tight")
```
